```python
X = digits.drop(['label'],axis=1)
Y = digits['label']
from sklearn.preprocessing import scale
X = scale(X)
```

```python
from sklearn.model_selection import train_test_split as tt
X_train , X_test , Y_train ,Y_test = tt(X,Y,test_size=0.3,random_state=100)
```

```python
print(X_test[1].shape)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
n_folds = KFold(n_splits = 10,shuffle= True ,random_state =100)
svm = SVC(C=1,kernel='linear', random_state =100)
accuracy = cross_val_score(svm ,X_train,Y_train,cv=n_folds,scoring ='accuracy')
print(accuracy)
print()
print(accuracy.mean())
```

```python
from sklearn.neighbors import KNeighborsClassifier
n_folds = KFold(n_splits = 10,shuffle= True ,random_state =100)
knn = KNeighborsClassifier()
accuracy = cross_val_score(knn ,X_train,Y_train,cv=n_folds,scoring ='accuracy')
print(accuracy)
print()
print(accuracy.mean())
```

```python
from sklearn.svm import SVC
n_folds = KFold(n_splits = 10,shuffle= True ,random_state =100)
svm = SVC(C=1,kernel='rbf', random_state =100)
accuracy = cross_val_score(svm ,X_train,Y_train,cv=n_folds,scoring ='accuracy')
print(accuracy)
print()
print(accuracy.mean())
```

```python
from sklearn.model_selection import GridSearchCV
model = SVC(kernel='rbf')
parameters = {'C': [0.1,1,10,100] , 'gamma' : [1e-1 ,1e-2 ,1e-3]}
grid_search = GridSearchCV(estimator=model,param_grid= parameters,scoring = 'accuracy')
grid_search.fit(X_train , Y_train)
print(grid_search.cv_results_)
```
